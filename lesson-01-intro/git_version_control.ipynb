# Claude Code 项目开发常用 Git 操作命令指南

## 一、克隆 Claude 项目到本地
```bash
git clone https://github.com/YourOrg/YourClaudeProject.git
cd YourClaudeProject
```

## 二、配置用户名与邮箱（首次操作需设置）
```
git config --global user.name "Your Name"
git config --global user.email "you@example.com"
```


## 三、创建与切换分支（推荐每个功能单独开分支）

### 3.1 新建功能分支（基于主分支 main）
```
git checkout main
git pull origin main
git checkout -b feat/add-agent-router
```

### 3.2 查看本地与远程分支
```
git branch           # 查看本地分支
git branch -r        # 查看远程分支
git branch -a        # 查看所有分支
```

## 四、提交代码流程

### 4.1 添加变更文件
```
git status                # 查看修改文件
git add .                 # 添加所有改动
```

```
# 或只添加某个文件：
git add prompts/router.claude.md
```

### 4.2 提交变更（规范备注）
```
git commit -m "feat: 添加任务路由Claude模块及Prompt设计"
```

## 五、推送分支到远程
```
git push origin feat/add-agent-router
```


## 六、从远程同步代码（防止冲突）

### 6.1 更新主分支
```
git checkout main
git pull origin main
```

### 6.2 更新当前开发分支（合并主分支最新代码）
```
git checkout feat/add-agent-router
git merge main
```


## 七、发布合并请求（Pull Request）

在 GitHub 网页点击 Compare & pull request 提交代码合并申请，并描述改动内容。


## 八、其他高级操作

### 8.1 删除本地分支

```
git branch -d feat/add-agent-router
```

### 8.2 删除远程分支
```
git push origin --delete feat/add-agent-router
```

### 8.3 撤回上一次提交但保留改动
```
git reset --soft HEAD~1
```
